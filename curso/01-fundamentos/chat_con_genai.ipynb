## Utilizando Gemini

La documentación se encuentra aquí: https://ai.google.dev/gemini-api/docs/migrate?hl=es-419

```python
from google import genai
from getpass import getpass
from google.genai import types
```

```python
GEMINI_API_KEY = getpass("Ingrese su API KEY...")
```

```python
client = genai.Client(api_key=GEMINI_API_KEY)
```

Generar contenido

```python
response = client.models.generate_content(
    model='gemini-flash-lite-latest',
    contents='Dime Hola en 3 idiomas diferentes.'
)
print(response.text)
```

Iniciar Chat

```python
chat = client.chats.create(model='gemini-flash-lite-latest')
```

```python
input_text = input("User: ")
response_chat = chat.send_message(
  message=input_text,
  config = types.GenerateContentConfig(
      system_instruction='Solamente en una sola línea y que alcance todo el contenido',
      max_output_tokens=20,
      temperature = 1
  )
)

count_tokens = client.models.count_tokens(
    model='gemini-flash-lite-latest',
    contents=response_chat.text
)

print(response_chat.text)
print(f"Tokens usados: {count_tokens.total_tokens}")
```

## Prompt

### Rol User

```python
def show_roles():
    print("="*50)
    print("Rol: User sin system")
    print("="*50)

    response = client.models.generate_content(
        model='gemini-flash-lite-latest',
        contents=[
            types.Content(
                role="user",
                parts=[types.Part(text="¿Cuánto es 2+2?")]
            )
        ]
    )

    print(f'resultado: {response.text}')

show_roles()
```

### Rol System

```python
def show_roles():
    print("="*50)
    print("Rol: User con system")
    print("="*50)

    response = client.models.generate_content(
        model='gemini-flash-lite-latest',
        contents=[
            types.Content(
                role="user",
                parts=[types.Part(text="¿Cuánto es 2+2?")]
            )
        ],
        config=types.GenerateContentConfig(
            system_instruction="""
            Eres un matemático gruñón que contesta preguntas con desgano,
            siempre incluyes un comentario sobre lo básico que es la pregunta.
            """
        )
    )

    print(f'resultado: {response.text}')

show_roles()
```

## Rol assistant

```python
def show_roles():
    print("="*50)
    print("Rol: Model clasificando comentario")
    print("="*50)

    response = client.models.generate_content(
        model='gemini-flash-lite-latest',
        contents=[
          types.Content(role="user",  parts=[types.Part(text="Clasifica: 'Me encantó'")]),
          types.Content(role="model", parts=[types.Part(text="POSITIVO")]),
          types.Content(role="user",  parts=[types.Part(text="Clasifica: 'Qué asco'")]),
          types.Content(role="model", parts=[types.Part(text="NEGATIVO")]),
          types.Content(role="user",  parts=[types.Part(text="Clasifica: 'Está bien'")]),
          types.Content(role="user",  parts=[types.Part(text="Me encantó este producto pero falta mejorar algunas cosas en sus entregas")])
        ],
        config=types.GenerateContentConfig(
            system_instruction="""
            Eres un clasificador de comentarios.
            Clasifica en: POSITIVO, NEGATIVO o NEUTRAL.
            Responde solamente con la clasificación asignada.
            """
        )
    )

    print(f'resultado: {response.text}')

show_roles()
```
